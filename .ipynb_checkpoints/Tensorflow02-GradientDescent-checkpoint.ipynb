### Tensorflow

`GradientDescentOptimizer`operation

Wb`1*1`

```python
import tensorflow as tf

#  y = W * x + b
# 
W = tf.Variable([0], dtype=tf.float32)  # 
b = tf.Variable([0.])                   # 

x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
y0 = W * x + b  # Wb

# xy
# 
loss = tf.reduce_sum(tf.square(y0 - y))

# rate0.01
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)
 
# W=[-1], b=[-1]
x_train = [-1, 6, 3, -4]
y_train = [0, -7, -4, 3]

# 
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)

    for i in range(301):
        sess.run(train, {x: x_train, y: y_train})
        if i % 20 == 0:
            print(sess.run([W, b, loss], {x: x_train, y: y_train}))
```

### 



$f'(x_0) = \lim\limits_{\Delta x\rightarrow 0} \frac{f(x_0 + \Delta x) - f(x_0)}{\Delta x}$f()

losslossf()f()

```python
import random

delta = 0.1 ** 5
rate = 0.1 ** 2

def f(x):
    return (x - 5) * (x - 2)

def loss(x):
    return f(x) ** 2

def gradient_descent(x):
    return x - rate * (loss(x + delta) - loss(x)) / delta

x = random.uniform(3, 4)
for i in range(101):
    x = gradient_descent(x)
    if i % 10 == 0:
        print('x:', x, ' loss:', loss(x))
```

```python
import random

delta = 0.1 ** 5
rate = 0.1 ** 2

def f(x, y):
    return (x - 5) ** 2 + (y - 2) ** 2

def loss(x, y):
    return f(x, y) ** 2

def gradient_descent(x, y):
    return [x - rate * (loss(x + delta, y) - loss(x, y)) / delta, 
            y - rate * (loss(x, y + delta) - loss(x, y)) / delta]

x = random.uniform(3, 4)
y = random.uniform(3, 4)
for i in range(401):
    x, y = gradient_descent(x, y)
    if i % 40 == 0:
        print('x:', x, 'y:', y, ' loss:', loss(x, y))
```
